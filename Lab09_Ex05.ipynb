## คำถาม

1. Static Method คืออะไร


คือเมธอดที่เป็นของคลาสไม่ใช่ของอ็อบเจ็กต์ เมธอดนี้จะถูกแชร์ระหว่างทุกอ็อบเจ็กต์ในคลาสนั้น ๆ และสามารถเรียกใช้ได้โดยไม่ต้องสร้างอ็อบเจ็กต์จากคลาส

2. ประกาศ static method ในภาษา C# อย่างไร


การประกาศ static method ใน C# ให้ใช้คีย์เวิร์ด static หน้าประกาศเมธอด

3. เรียกใช้ static method อย่างไร


สามารถเรียกใช้ static method ได้โดยใช้ชื่อคลาส ตามด้วยชื่อเมธอด

4. static method สามารถเข้าถึง instance members (fields, methods) ได้หรือไม่


- ไม่ได้ เพราะ Static method ไม่สามารถเข้าถึง instance members (เช่น instance fields หรือ instance methods) โดยตรงได้ เนื่องจาก static method ไม่เกี่ยวข้องกับอ็อบเจ็กต์ใด ๆ ของคลาส
- ถ้าต้องการเข้าถึง instance members ใน static method ต้องสร้างอ็อบเจ็กต์ของคลาสนั้นขึ้นมาก่อนแล้วเข้าถึง instance members ผ่านอ็อบเจ็กต์

5. static method เหมาะกับการใช้งานแบบไหน


- Static method เหมาะสำหรับการใช้งานที่ไม่ต้องการการเข้าถึงหรือเปลี่ยนแปลงสถานะของอ็อบเจ็กต์ (instance) เช่น ฟังก์ชันคำนวณค่าคงที่ หรือฟังก์ชันที่ทำงานกับข้อมูลที่เป็น static (ที่ใช้ร่วมกันระหว่างอ็อบเจ็กต์)

6. ข้อดีของการใช้ static method คืออะไรบ้าง


- ประหยัดหน่วยความจำ: เนื่องจาก static method ใช้ร่วมกันทั้งหมดในคลาสเดียวกัน จึงไม่ต้องสร้างอ็อบเจ็กต์ใหม่เพื่อเรียกใช้เมธอดนั้น ๆ
- ง่ายต่อการเรียกใช้งาน: เนื่องจากสามารถเรียกใช้ได้โดยไม่ต้องสร้างอ็อบเจ็กต์

7. ข้อเสียของการใช้ static method คืออะไรบ้าง


- จำกัดการใช้งาน: Static method ไม่สามารถเข้าถึง instance members (fields หรือ methods) ได้, ทำให้ไม่สามารถใช้ในกรณีที่ต้องการทำงานร่วมกับข้อมูลที่เฉพาะเจาะจงกับอ็อบเจ็กต์
- ยากในการทดสอบ: การใช้ static method อาจทำให้ทดสอบโค้ดยากขึ้น เนื่องจากเมธอดเหล่านี้อาจเกี่ยวข้องกับสถานะที่ร่วมกันในคลาส
- ลดความยืดหยุ่น: การใช้ static method มากเกินไปอาจทำให้โค้ดขาดความยืดหยุ่นในการขยายและปรับปรุง
