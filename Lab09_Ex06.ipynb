## คำถาม

1. Member constant คืออะไร และมีคุณสมบัติอย่างไร


 - คือค่าคงที่ที่ประกาศภายในคลาสหรือโครงสร้างในภาษา C# ซึ่งมีค่าคงที่และไม่สามารถเปลี่ยนแปลงได้ระหว่างการทำงานของโปรแกรม
 - คุณสมบัติ:
ค่าคงที่ที่ประกาศด้วย const จะถูกกำหนดค่าไว้ในขณะคอมไพล์ และไม่สามารถเปลี่ยนแปลงค่าได้หลังจากที่คอมไพล์เสร็จ
ค่าคงที่นั้นใช้ร่วมกันได้ระหว่างทุกอ็อบเจ็กต์ของคลาส
ถูกใช้สำหรับเก็บข้อมูลที่ไม่เปลี่ยนแปลงตลอดเวลา เช่น ค่า PI หรือค่าคงที่ทางคณิตศาสตร์

2. เราสามารถประกาศ member constant ในภาษา C# ได้อย่างไร


- การประกาศ member constant ใน C# ให้ใช้คีย์เวิร์ด const ตามด้วยประเภทข้อมูลและชื่อของค่าคงที่ 

````
class Circle
{
    public const double PI = 3.14159;
}
````

3. member constant สามารถสร้างเป็นชนิดข้อมูลใดได้บ้าง


- ตัวเลข เช่น int, double, float, decimal
- ข้อความ เช่น string
- ค่า Boolean เช่น bool

4. เราสามารถเข้าถึง member constant ได้อย่างไร


สามารถเข้าถึงได้ผ่านชื่อคลาส โดยไม่จำเป็นต้องสร้างอ็อบเจ็กต์ของคลาส

````
Console.WriteLine(Circle.PI);  // เข้าถึงค่าคงที่ PI ในคลาส Circle
````

5. member constant เป็น static โดยปริยายหรือไม่


- ใช่: ใน C#, member constant จะเป็น static โดยปริยาย เพราะมันเป็นค่าคงที่ที่ใช้ร่วมกันระหว่างอ็อบเจ็กต์ทั้งหมดในคลาสเดียวกัน

6. มีข้อจำกัดอะไรบ้างในการใช้ member constant 


- ไม่สามารถเปลี่ยนค่าได้: ค่าที่ประกาศเป็น const จะไม่สามารถเปลี่ยนแปลงค่าได้หลังจากคอมไพล์
- ต้องกำหนดค่าในขณะประกาศ: เมื่อประกาศ member constant จะต้องกำหนดค่าให้เสร็จสิ้นในขณะคอมไพล์
- ไม่สามารถใช้กับชนิดข้อมูลที่ไม่สามารถคำนวณในขณะคอมไพล์ได้: เช่น DateTime, dynamic, หรือ object ที่ค่าไม่สามารถคำนวณได้ในขณะคอมไพล์

7. member constant เหมาะกับการใช้งานแบบไหน


- การเก็บค่าคงที่ที่ไม่เปลี่ยนแปลง: เหมาะสำหรับการเก็บข้อมูลที่ไม่ควรเปลี่ยนแปลง เช่น ค่าคงที่ทางคณิตศาสตร์ (PI), ค่าคงที่ในโปรแกรม (MaxUsers), หรือค่าคงที่ในระบบ (MaxRetries)
- เพิ่มความปลอดภัย: ค่าคงที่ช่วยป้องกันการเปลี่ยนแปลงที่ไม่ได้ตั้งใจของข้อมูลที่สำคัญในโปรแกรม
