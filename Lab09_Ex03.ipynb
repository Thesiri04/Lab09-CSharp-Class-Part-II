## คำถาม

1. ในภาษา C# นั้น Instance members และ static members  แตกต่างกันอย่างไร



- **Instance Members**:  
  - ต้องเข้าถึงผ่านอ็อบเจ็กต์  
  - ข้อมูลแยกกันสำหรับแต่ละอ็อบเจ็กต์  
  - ใช้เมื่อข้อมูลหรือพฤติกรรมต้องแยกต่างกันในแต่ละอ็อบเจ็กต์

- **Static Members**:  
  - เข้าถึงได้ผ่านชื่อคลาสโดยตรง  
  - ข้อมูลใช้ร่วมกันในทุกอ็อบเจ็กต์ของคลาส  
  - ใช้เมื่อข้อมูลหรือฟังก์ชันต้องใช้ร่วมกันทั่วทั้งคลาส



2. เข้าถึง Instance members และ static members มีหลักการที่แตกต่างกันอย่างไร

1. Instance Members
- ต้องสร้างอ็อบเจ็กต์ของคลาสก่อน เพื่อที่จะเข้าถึงฟิลด์หรือเมธอดที่เป็น instance
- ใช้การเข้าถึงผ่านตัวแปรอ็อบเจ็กต์ (เช่น objectName.member)
2. Static Members
- ไม่ต้องสร้างอ็อบเจ็กต์ ของคลาสก่อน
- เข้าถึงได้โดยตรงผ่านชื่อคลาส (เช่น ClassName.StaticMember)
- ข้อมูลหรือฟังก์ชันที่เป็น static จะใช้ร่วมกันในทุกอ็อบเจ็กต์ของคลาสเดียวกัน


3. Instance members และ static members เหมาะกับการใช้งานแบบไหน

1. Static Members
เหมาะสำหรับการใช้งานเมื่อ

- ต้องการให้ข้อมูลหรือพฤติกรรมใช้ร่วมกันในทุกอ็อบเจ็กต์ของคลาสเดียวกัน
- ข้อมูลหรือฟังก์ชันที่ไม่ขึ้นอยู่กับสถานะของอ็อบเจ็กต์
- การเก็บข้อมูลหรือฟังก์ชันที่มีค่าร่วมกัน เช่น ค่าคงที่ (constants), ตัวนับ (counters), ฟังก์ชันช่วยที่ไม่เกี่ยวข้องกับอ็อบเจ็กต์เฉพาะ
2. Instance Members
เหมาะสำหรับการใช้งานเมื่อ:

- ต้องการให้ข้อมูลหรือพฤติกรรมมีค่าแยกกันสำหรับแต่ละอ็อบเจ็กต์
- ข้อมูลในแต่ละอ็อบเจ็กต์ไม่สัมพันธ์กันและต้องเก็บแยกต่างหาก
- เมื่ออ็อบเจ็กต์แต่ละตัวมีสถานะหรือพฤติกรรมที่ไม่เหมือนกัน

4. Instance members และ static members ใช้หน่วยความจำต่างกันอย่างไร

1. Instance members จะถูกจัดเก็บแยกต่างหากสำหรับแต่ละอ็อบเจ็กต์ของคลาส
2. Static members จะถูกจัดสรรหน่วยความจำเพียงครั้งเดียวสำหรับคลาสนั้น ๆ และใช้ร่วมกันในทุกอ็อบเจ็กต์ของคลาส

5. Instance members และ static members มีอายุการใช้งานต่างกันอย่างไร

1. Instance members จะมีอายุการใช้งานที่ผูกกับอ็อบเจ็กต์ของคลาส พวกมันจะ อยู่ในหน่วยความจำ ตราบเท่าที่อ็อบเจ็กต์นั้น ๆ ยังมีชีวิตอยู่
2. Static members จะมีอายุการใช้งานที่ผูกกับ คลาส แทนที่จะเป็นอ็อบเจ็กต์
พวกมันจะ อยู่ในหน่วยความจำ ตราบเท่าที่โปรแกรมทำงานอยู่ หรือจนกว่าคลาสนั้นจะถูกยกเลิกการใช้งาน

6. Instance members และ static members สามารถใช้งานร่วมกันได้หรือไม่

- ทั้ง Instance members และ Static members สามารถใช้งานร่วมกันได้ในบางกรณี แต่มีข้อแตกต่างในการเข้าถึงและการใช้งานที่ควรระวัง

7. keyword static ใช้กับสมาชิกประเภทใดของคลาสได้บ้าง

- Static Fields: ตัวแปรที่ใช้ร่วมกันในทุกอ็อบเจ็กต์
- Static Methods: เมธอดที่ไม่ต้องการการสร้างอ็อบเจ็กต์
- Static Properties: พร็อพเพอร์ตี้ที่ใช้ร่วมกันในทุกอ็อบเจ็กต์
- Static Constructors: ตัวสร้างที่ใช้สำหรับการกำหนดค่าเริ่มต้น
- Static Indexers: อินเด็กซ์ที่ใช้ร่วมกันในคลาส
